```python
import pandas as pd
import numpy as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris_df=sns.load_dataset('iris')
iris_df
```

```python
iris_df['species'].value_counts()
```

```python
sns.scatterplot(x=iris_df['sepal_length'],y=iris_df['petal_length'],color='red')
plt.show()
```

```python
sns.scatterplot(x=iris_df['sepal_length'],y=iris_df['petal_length'],hue=iris_df['species'])
plt.show()
```

```python
sns.boxplot(x=iris_df['species'],y=iris_df['petal_length'])
plt.show()
```

```python
sns.violinplot(x=iris_df['species'],y=iris_df['sepal_width'])
plt.show()
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'])
plt.show() #combinations of 2 plots below are types of joints
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'] ,kind='reg')
plt.show() 
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'] ,kind='scatter')
plt.show() 
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'] ,kind='kde')
plt.show() 
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'] ,kind='hex')
plt.show() 
```

```python
sns.jointplot(x=iris_df['sepal_length'],y=iris_df['sepal_width'] ,kind='hex',color='red')
plt.show() 
```

```python
sns.countplot(x=iris_df['sepal_length'],hue=iris_df['species'])
plt.show()
```

```python
sns.countplot(x=iris_df['species'])
plt.show()
```

```python
sns.countplot(y=iris_df['species'],color='purple')
plt.show()
```

```python
sns.countplot(y=iris_df['species'],palette=sns.cubehelix_palette())
plt.show()
```
